```python
def average_Statistics(dataset):
    mean = np.mean(dataset)
    median = np.median(dataset)

    return mean, median
```

```python
def spread_Statistic(dataset):
    std = np.std(dataset)
    var = np.var(dataset)
    
    return std, var
```

```python
def print_Statistics(name1, value1, name2, value2, unit = "", dataName = "none"):
    if dataName != "none":
        print("\n", dataName, ":")
        
    print("\t", name1, ":", np.round(value1, decimals=3), unit)
    print("\t", name2, ":", np.round(value2, decimals=3), unit)
```

```python
        ### Using the iris data set from Scikit-learn ###

# https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html

from sklearn.datasets import load_iris
iris_sklearn = load_iris()
X = iris_sklearn.data
y = iris_sklearn.target
```

```python
# 1. Build a decision tree classifier using scikit-learn's DecisionTreeClassifier
# (using the default options)

# https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html
# https://scikit-learn.org/stable/modules/tree.html

from sklearn.tree import DecisionTreeClassifier
tree_classifier = DecisionTreeClassifier(random_state=0)
tree_classifier = tree_classifier.fit(X, y)
```

```python
# 2. Plot the resulting decision tree.

# https://chrisalbon.com/machine_learning/trees_and_forests/visualize_a_decision_tree/

from IPython.display import Image  
from sklearn import tree
import pydotplus

# Create DOT data
dot_data = tree.export_graphviz(tree_classifier, out_file=None, 
                                feature_names=iris_sklearn.feature_names,  
                                class_names=iris_sklearn.target_names)

# Draw graph
graph = pydotplus.graph_from_dot_data(dot_data)  

# Show graph
Image(graph.create_png())
```

```python
            ### Using the iris dataset from Seaborn ###

# Splitting the dataframe to data and target(label)
data = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
target = iris[['species']]

# Build a decision tree classifier
tree_classifier = DecisionTreeClassifier(random_state=0)
tree_classifier = tree_classifier.fit(data, target)

# Plot the decision tree
dot_data = tree.export_graphviz(tree_classifier)
graph = pydotplus.graph_from_dot_data(dot_data)  
Image(graph.create_png())
```

```python
# 3. Perform k-fold cross-validation using k=3 and display the results

# https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html
# cv = Determines the cross-validation splitting strategy. Default 3-fold cross validation

from sklearn.model_selection import cross_val_score

default_scores = cross_val_score(tree_classifier, X, y, cv=3)
print("Cross-validation Scores:",', '.join(map(str, np.round(default_scores,decimals=5))))
print("Average of Cross-validation Scores : ", np.round(np.mean(default_scores),decimals=5))
```

In this part you will build a face recognition solution.

### Write code to: 

1. Use Principal Component Analysis (PCA) to reduce the dimensionality of each face to the first 120 components. (10 pts)
2. Build and fit a multi-class SVM classifier, with sensible initialization options, and using the PCA-reduced  features. (10 pts)
3. Make predictions on the test data, compute the precision, recall and f1 score for each category, compute the overall accuracy, and plot the resulting confusing matrix. (25 pts)
4. Display examples of correct and incorrect predictions (at least 5 of each). (10 pts)

## Solution

```python
# 1. Use Principal Component Analysis (PCA) to reduce the 
# dimensionality of each face to the first 120 components

# https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html
# https://scikit-learn.org/stable/modules/generated/sklearn.datasets.fetch_lfw_people.html
# 05.07-Support-Vector-Machines.ipynb

# faces.data = Each row corresponds to a ravelled face image of original size 62 x 47 pixels

from sklearn.decomposition import PCA

face_pca = PCA(n_components=120, whiten=True, random_state=42)
face_pca = face_pca.fit(faces.data)
components = face_pca.transform(faces.data)

print("Original shape: ", faces.data.shape)
print("PCA Dimension Reduction:",components.shape)
```

```python
# Splitting the dataset to train and test set 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(components, faces.target,
                                                    random_state=0)
```

```python
# 2. Build and fit a multi-class SVM classifier, with sensible
# initialization options, and using the PCA-reduced features. 

# https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html

from sklearn.svm import SVC

face_svc_model = SVC(kernel='rbf', class_weight='balanced')
face_svc_model.fit(X_train, y_train)
```

```python
from sklearn.feature_extraction.image import PatchExtractor

def extract_patches(img, N, scale=1.0, patch_size=positive_patches[0].shape):
    extracted_patch_size = tuple((scale * np.array(patch_size)).astype(int))
    extractor = PatchExtractor(patch_size=extracted_patch_size,
                               max_patches=N, random_state=0)
    patches = extractor.transform(img[np.newaxis])
    if scale != 1:
        patches = np.array([transform.resize(patch, patch_size)
                            for patch in patches])
    return patches

negative_patches = np.vstack([extract_patches(im, 1000, scale)
                              for im in images for scale in [0.5, 1.0, 2.0]])
negative_patches.shape
```

```python
def sliding_window(img, patch_size=positive_patches[0].shape,
                   istep=2, jstep=2, scale=1.0):
    Ni, Nj = (int(scale * s) for s in patch_size)
    for i in range(0, img.shape[0] - Ni, istep):
        for j in range(0, img.shape[1] - Ni, jstep):
            patch = img[i:i + Ni, j:j + Nj]
            if scale != 1:
                patch = transform.resize(patch, patch_size)
            yield (i, j), patch
            
indices, patches = zip(*sliding_window(test_image))
patches_hog = np.array([feature.hog(patch) for patch in patches])
patches_hog.shape
```

### Lecture from Coursera by Andrew Ng

https://www.coursera.org/lecture/convolutional-neural-networks/non-max-suppression-dvrjH

Steps for non-maximal suppression:
1. Discard any box that is the prediction chance < 0.60.
2. Pick the bounding box that has the highest prediction.
3. Discard any boxes that has a high overlap on box in Step 2.

In the given code from the textbook, it only give probably of either 0 or 1 so I don't think this way will work with have we have. As the result, I came up the a modified version that is less robust but work in this case:
1. Use the x value that has the highest frequency count. Higher count giving higher priority since that mean the boxes are overlapping there and there is a likelyhood of a detection. 
2. Calculate for the mean value of y coordinates that belong to x in Step 1.
3. Save the coordinates in the resulting array. 
4. After the first saved value, other preceding detection need to be check with the threshold value (that the user set) to see if we need to save more coordinates. 

```python
### Improve it to include non-maximal suppression ### 

yx_Values = indices[labels == 1]
yx_Values 
```

```python
y_coordinates, x_coordinates = yx_Values [:,0], yx_Values[:,1]
print(x_coordinates)
print(y_coordinates)
```

```python
def MODIFIED_nonMaximalSuppression(yx_Values,threshold = 10):
    y_coordinates, x_coordinates = yx_Values[:,0], yx_Values[:,1]
    
    count = 0;
    result_array = np.empty([1, 2])
    
    while (x_coordinates.size != 0):
        # Most frequent x value
        x_counts = np.bincount(x_coordinates)
        frequent_x_value = np.argmax(x_counts)
        
        # x_indexes for the most frequent x value
        x_indexes = np.where(x_coordinates == frequent_x_value)

        # Mean of y values at the most frequent x value
        y_mean = np.mean(y_coordinates[x_indexes])
        y_mean = int(round(y_mean))
        
        # Delete the frequent values in arrays
        x_coordinates = np.delete(x_coordinates, x_indexes)
        y_coordinates = np.delete(y_coordinates, x_indexes)
        
        # Check the threshold to see if we need to save more coordinates 
        # if the frequent x values is > largest or < smallest of the 
        # coordinates in the resulting list
        if count == 0:
            result_array = np.array([frequent_x_value, y_mean])
        else:
            if (result_array.size == 2):
                result_x_values = result_array[0]
            else:
                result_x_values = result_array[:,0]
                
            min_x_values = np.amin(result_x_values)
            max_x_values = np.amax(result_x_values)
            
            if ((min_x_values - threshold) > frequent_x_value) or ((max_x_values + threshold) < frequent_x_value):
                additional_result =np.array([frequent_x_value, y_mean])
                result_array = np.vstack((result_array, additional_result))
        
        count = count + 1
    
    return result_array
```
